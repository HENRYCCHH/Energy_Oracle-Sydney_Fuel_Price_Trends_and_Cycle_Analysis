```python
import pandas as pd

# We'll read sheets one by one to avoid memory issues
combined_df_parts = []

for file_path in [
    '/Users/henrychiu/Documents/fuel check data/2022/combined_fuel_price_history_2022_cleaned.csv',
    '/Users/henrychiu/Documents/fuel check data/2023/combined_fuel_price_history_2023_cleaned.csv',
    '/Users/henrychiu/Documents/fuel check data/2024/combined_fuel_price_history_2024_cleaned.csv'
]:
    # Read only the first sheet to keep it consistent and fast
    df = pd.read_csv(file_path)
    combined_df_parts.append(df)

# Combine all dataframes
combined_df = pd.concat(combined_df_parts, ignore_index=True)

combined_df.head()
```

```python
# Convert 'PriceUpdatedDate' to datetime format
combined_df['PriceUpdatedDate'] = pd.to_datetime(combined_df['PriceUpdatedDate'], errors='coerce')
print(combined_df['PriceUpdatedDate'].dtypes)
```

```python
# Save combined file
output_path = "/Users/henrychiu/Documents/fuel check project/fuel check data/combined_fuel_price_history_2022_2023_2024_cleaned_v2.csv"
combined_df.to_csv(output_path, index=False)
```
